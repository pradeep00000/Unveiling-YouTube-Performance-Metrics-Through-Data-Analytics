```python
df = df.dropna().copy()
df
```

```python
df.drop(columns=['ID'],inplace=True)
```

```python
# Distribution of views
plt.figure(figsize=(6,3))
sns.histplot(df['Views'], bins=30, kde=True)
plt.title('Distribution of Video Views')
plt.show()
```

```python
# Corrected code for a better correlation heatmap
plt.figure(figsize=(12,7))  # Adjusted the figure size for better visibility
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', annot=True, fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap with Annotations')
plt.show()
```

```python
# Average revenue by day of the week
plt.figure(figsize=(6,3))
avg_revenue_per_day = df.groupby('Day of Week')['Estimated Revenue (USD)'].mean()
avg_revenue_per_day.plot(kind='bar', color='skyblue')
plt.title('Average Revenue by Day of the Week')
plt.ylabel('Revenue (USD)')
plt.xlabel('Day of the Week')
plt.show()
```

```python
# Scatter plot for Revenue vs. Views
plt.figure(figsize=(6,4))
sns.scatterplot(x='Views', y='Estimated Revenue (USD)', data=df)
plt.title('Relationship Between Views and Revenue')
plt.xlabel('Views')
plt.ylabel('Estimated Revenue (USD)')
plt.show()
```

```python
# Analyzing the relationship between average view duration and watch time
sns.lineplot(x='Average View Duration', y='Watch Time (hours)', data=df)
plt.title('Average View Duration vs Watch Time')
plt.show()
```

Conclusion:
From the analysis, we found significant correlations between views and estimated revenue, suggesting that higher views generally result in higher revenue. We also observed which days tend to generate higher average revenue, offering potential insights for content upload strategies. The scatter plot between views and revenue shows a positive relationship but with variability that may imply other contributing factors.
